```python
import SimpleITK as sitk
import os
import numpy as np
```

```python
model_directory = "C:/Users/lydia/OneDrive/Work/DigitalAssets/_UPENN_fMOST/"
model_file = os.path.join( model_directory, "25um_V3", "CCFTemplate25umLabels.nii.gz")
```

```python
model = sitk.ReadImage( model_file )
arr = sitk.GetArrayFromImage( model )
unique_values  = np.unique( arr )

# initialization
all_boundaries = None

for a in unique_values :

    if a > 0 :

        print("label: %d" % a )

        mask = sitk.Equal( model, int(a), 0, 1 )
        boundary = sitk.BinaryContour( mask, True, 0, 1 )

        if all_boundaries :
            all_boundaries = sitk.Maximum( boundary, all_boundaries )
        else :
            all_boundaries = sitk.Maximum( boundary, boundary )
```

```python
output_file = os.path.join( model_directory, "25um_V3", "CCFTemplate25umBoundaries.nii.gz")
sitk.WriteImage( all_boundaries, output_file, True )
```
