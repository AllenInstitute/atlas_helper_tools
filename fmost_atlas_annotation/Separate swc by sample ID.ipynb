```python
import os
import glob
import shutil
```

```python
model_directory = "C:/Users/lydia/OneDrive/Work/DigitalAssets/_UPENN_fMOST"

input_directory = os.path.join( model_directory,"HUSTReproduction_9_7_2021","Original","HUST_REGISTRATION_AllCells_OriginalSent")
output_directory = os.path.join( model_directory,"HUSTReproduction_9_7_2021","Sandbox")
```

```python
slist = []

for ipath in glob.glob(os.path.join(input_directory, "*.swc")):
    
    print(ipath)
    
    bname = os.path.basename(ipath)
    sname = os.path.splitext(bname)[0]
    sname = sname.split('_')[0]
    
    if not (sname in slist) :
        slist.append(sname)
    
    odir = os.path.join( output_directory, "HUST_REGISTRATION_AllCells_OriginalSent", sname )
    
    if not os.path.exists(odir):
        os.makedirs(odir)
    
    opath = os.path.join(odir, bname)
    shutil.copyfile( ipath, opath )
    
```

```python
steps = []
steps.append('HUST_REGISTRATION_AllCells_OriginalSent_thumbnail')
steps.append('HUST_REGISTRATION_AllCells_in_micron')
steps.append('HUST_REGISTRATION_AllCells_in_micron_registered')
steps.append('HUST_REGISTRATION_AllCells_in_micron_registered_thumbanil')

for t in steps:
    for s in slist:
        odir = os.path.join( output_directory, t, s )
        if not os.path.exists(odir):
            os.makedirs(odir)
```
