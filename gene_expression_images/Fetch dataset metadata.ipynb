```python
import urllib, json
import os
import pandas as pd
```

```python
#
# retrieve metdata in batches 
#

def fetch_batch( start_row = 0, num_row = 100, product_id = 3 ) :
    
    query_url = \
    "http://api.brain-map.org/api/v2/data/query.json?criteria=model::SectionDataSet,rma::criteria,[failed$eqfalse],products[id$eq%d],rma::include,specimen(donor(age)),treatments,genes,products,plane_of_section,rma::options[only$eq'data_sets.id,products.name,data_sets.section_thickness,plane_of_sections.name,specimens.id,specimens.name,donors.sex_full_name,ages.name,treatments.name,genes.acronym'][start_row$eq%d][num_rows$eq%d]" \
    % (product_id,start_row,num_row)

    response = urllib.request.urlopen(query_url)
    output = json.loads(response.read())['msg']
    
    df = pd.DataFrame()
    df['section_dataset_id'] = [ds['id'] for ds in output ]
    df['section_thickness'] = [ds['section_thickness'] for ds in output ]
    df['plane_of_section'] = [ds['plane_of_section']['name'] for ds in output ]
    df['product_name'] = [ds['products'][0]['name'] for ds in output ]
    df['specimen_id'] = [ds['specimen']['id'] for ds in output ]
    df['specimen_name'] = [ds['specimen']['name'] for ds in output ]
    df['age'] = [ds['specimen']['donor']['age']['name'] for ds in output ]
    df['treatment'] = [ds['treatments'][0]['name'] for ds in output ]
    df['gene_symbol'] = [ds['genes'][0]['acronym'] if (len(ds['genes'])>0) else None for ds in output]
    
    return df
```

```python
pbatches = {}
pbatches[3] = 218
pbatches[4] = 1
nrow = 100
df = pd.DataFrame()

for p in pbatches:
    
    for b in range(pbatches[p]) :
        
        sr = b * nrow
        
        print((p,sr))
        
        bdf = fetch_batch(sr,nrow,p)
        
        df = pd.concat([df,bdf],ignore_index=True)
        
df = df.drop_duplicates()
```
