```python
import gene_expression_helper as helper
import alignment_helper as algn
import pandas as pd
import os
```

```python
base_directory = r"\\allen\aibs\ccf\mouse_gene_expression\202201121424_batch"
```

```python
manifest_file = os.path.join( base_directory, 'manifest.csv')
df = pd.read_csv( manifest_file )

if 0 :
    for index, row in df.iterrows() :
        dname = row['directory_name']
        arr = dname.split('-')
        df.loc[index,'gene_symbol'] = arr[0]
        df.loc[index,'dataset_id'] = arr[1]

    df.to_csv( manifest_file, index=False )
```

```python
product_id = 1
downsample_factor = 3

atlas_directory =  r'C:\Users\LydiaN\OneDrive - Personal\OneDrive\Work\DigitalAssets\ccf_models'
```

```python
# read atlas
atlas = algn.read_atlas( atlas_directory, 'P56', 'coronal')

for index, row in df.iterrows() :
    
    print(row['directory_name'])
    
    dataset_directory = os.path.join( base_directory, row['directory_name'])
    os.makedirs( dataset_directory, exist_ok = True )
    
    dataset_id = row['dataset_id']
    
    # download metadata
    metadata_json = os.path.join(dataset_directory, "metadata_%d.json" % (dataset_id) )
    helper.download_metadata( dataset_id, metadata_json, product_id )
    
    # download downsampled ISH images
    image_directory = os.path.join( dataset_directory, 'section_images' )
    os.makedirs( image_directory, exist_ok = True )
    helper.download_images(dataset_id,image_directory,downsample_factor)
    
    # download expression images
    expression_directory = os.path.join( dataset_directory, 'expression_images' )
    os.makedirs( expression_directory, exist_ok = True )
    helper.download_expression_images(dataset_id,expression_directory,downsample_factor)
    
    # download alignment parameters
    alignment_json = os.path.join(dataset_directory, "alignment_%d.json" % (dataset_id) )
    helper.download_alignment_parameters( dataset_id, alignment_json, product_id )
    
    # Use the alignment parameters to create reconstructed and registered volumes
    xform = algn.read_json( alignment_json)

    # generate reconstructed volume
    output = algn.initialize_output( xform, atlas_directory, downsample_factor )
    algn.populate_output( xform, image_directory, downsample_factor, output, algn.GREENCHANNEL, expression_directory )
    algn.fill_in_missing( output )
    
    # resample to atlas
    algn.resample_to_atlas( xform, atlas, output )

    # write volume to output
    resampled_directory = os.path.join( dataset_directory, 'resampled')
    os.makedirs( resampled_directory, exist_ok = True )
    algn.write_volumes( output, resampled_directory, only = ['resampled_volume','resampled_expression'] )
    
    
```
